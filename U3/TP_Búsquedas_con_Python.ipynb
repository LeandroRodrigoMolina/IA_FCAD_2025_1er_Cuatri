### **N-reinas***

El problema de las N-Reinas es un clásico problema en teoría de la computación. Consiste en ubicar N-reinas en un tablero de ajedrez de N×N casillas, de forma tal que ninguna reina ataque a otra.

image.png

 A partir del siguiente código *Python* donde se presenta una solución para el problema de las n-reinas.



1.   Adecuar la implementación para n-torres.


```python
"""Puzzle n-reinas."""
class NQueens:
    """Genere todas las soluciones válidas para el puzzle de n reinas"""
    def __init__(self, size):
        # Almacene el tamaño del puzzle (problema) y la cantidad de soluciones válidas
        self.size = size
        self.solutions = 0
        self.solve()

    def solve(self):
        """Resuelve el puzzle de las n reinas e imprime el número de soluciones"""
        positions = [-1] * self.size
        self.put_queen(positions, 0)
        print("Found", self.solutions, "solutions.")

    def put_queen(self, positions, target_row):
        """
        Intente colocar una reina en target_row marcando todos los N casos posibles.
        Si se encuentra un lugar válido, la función se llama a sí misma tratando de colocar
        una reina en la siguiente fila hasta que todas las N reinas se colocan en el tablero NxN.
        """
        # Caso base (corte): todas las N filas están ocupadas
        if target_row == self.size:
            self.show_full_board(positions)
            # self.show_short_board(positions)
            self.solutions += 1
        else:
            # Para todas las posiciones de N columnas, intente colocar una reina
            for column in range(self.size):
                # Rechazar todas las posiciones inválidas
                if self.check_place(positions, target_row, column):
                    positions[target_row] = column
                    self.put_queen(positions, target_row + 1)


    def check_place(self, positions, ocuppied_rows, column):
        """
        Compruebe si una posición determinada está siendo atacada
        por alguna de las reinas colocadas anteriormente
        (verifique las posiciones de la columna y la diagonal)
        """
        for i in range(ocuppied_rows):
            if positions[i] == column:
                return False
        return True

    def show_full_board(self, positions):
        """Mostrar el tablero completo de NxN"""
        for row in range(self.size):
            line = ""
            for column in range(self.size):
                if positions[row] == column:
                    line += "Q "
                else:
                    line += ". "
            print(line)
        print("\n")

    def show_short_board(self, positions):
        """
        Muestre las posiciones de las reinas en el tablero en forma comprimida,
        cada número representa la posición de la columna ocupada en la fila correspondiente.
        """
        line = ""
        for i in range(self.size):
            line += str(positions[i]) + " "
        print(line)

def main():
    """Inicializa y resuelve el rompecabezas de n reinas"""
    NQueens(4)

if __name__ == "__main__":
    # execute only if run as a script
    main()
```

```python
from simpleai.search import astar, SearchProblem
# Clase que contiene métodos para resolver el puzzle.
class PuzzleSolver(SearchProblem):
    #
    # Explicar el objetivo de "def actions(self, cur_state)"
    # Esta funcion determina las acciones posibles a partir del estado actual
    # y devuelve una lista de acciones posibles en este caso los números que se pueden mover al espacio vacío
    # Identifica que piezas se pueden mover al espacio vacio, la de arriba, abajo, izquierda y derecha
    # 
    def actions(self, cur_state):
        rows = string_to_list(cur_state)
        row_empty, col_empty = get_location(rows, 'z')

        actions = []
        if row_empty > 0:
            actions.append(rows[row_empty - 1][col_empty])
        if row_empty < 2:
            actions.append(rows[row_empty + 1][col_empty])
        if col_empty > 0:
            #
            # Completar con el código correspondiente
            #
            actions.append(rows[row_empty][col_empty - 1])
        if col_empty < 2:
            #
            # Completar con el código correspondiente
            #
            actions.append(rows[row_empty][col_empty + 1])
        return actions

    # Devuelve el estado resultante después de mover una pieza al espacio vacío
    def result(self, state, action):
        rows = string_to_list(state)
        row_empty, col_empty = get_location(rows, 'z')
        row_new, col_new = get_location(rows, action)

        rows[row_empty][col_empty], rows[row_new][col_new] = \
                rows[row_new][col_new], rows[row_empty][col_empty]

        return list_to_string(rows)

    # Retorna verdadero si el estado es estado objetivo
    def is_goal(self, state):
        return state == GOAL # Completar con el código correspondiente

    #
    # Explicar el objetivo de "def heuristic(self, state)"
    # Indicar que heurística se utiliza
    # Usa la distancia Manhattan para cada pieza en el estado actual
    # y la posición de la pieza en el estado objetivo
    # retorna suma de la distancia entre la posicion actual y la posicion objetivo
    def heuristic(self, state):
        rows = string_to_list(state)

        distance = 0

        for number in '12345678z':
            row_new, col_new = get_location(rows, number)
            row_new_goal, col_new_goal = goal_positions[number]

            distance += abs(row_new - row_new_goal) + abs(col_new - col_new_goal)

        return distance

# Convierte lista en string
def list_to_string(input_list):
    return '\n'.join(['-'.join(x) for x in input_list])

# Convierte string en lista
def string_to_list(input_string):
    return [x.split('-') for x in input_string.split('\n')]

# Encuentra la ubicación 2D del elemento de entrada
def get_location(rows, input_element):
    for i, row in enumerate(rows):
        for j, item in enumerate(row):
            if item == input_element:
                return i, j

# Final result that we want to achieve
GOAL = '''1-2-3
4-5-6
7-8-z'''

# Starting point
INITIAL = '''1-z-2
6-5-4
8-7-3'''
##
## Completar con el código correspondiente
##

#Crea un caché para la posición de la meta de cada pieza.
goal_positions = {}
rows_goal = string_to_list(GOAL)
for number in '12345678z':
    goal_positions[number] = get_location(rows_goal, number)

#
# Explicar el objetivo de "result"
# Ejecuta el algoritmo A* para resolver el puzzle desde el estado inicial
# y devuelve la secuencia de movimientos que llevan al estado objetivo
#
result = astar(PuzzleSolver(INITIAL))

# Muestra en pantalla los resultados
for i, (action, state) in enumerate(result.path()):
    print()
    if action == None:
        print('Configuración inicial')
    elif i == len(result.path()) - 1:
        print('Después de mover', action, 'al espacio vacío. Objetivo alcanzado!')
    else:
        print('Después de mover', action, 'al espacio vacío')

    print(state)
```

```python
from simpleai.search import CspProblem, backtrack
import matplotlib.pyplot as plt
import networkx as nx

departamentos = [
    'Paraná', 'Diamante', 'Victoria', 'Gualeguay', 'Gualeguaychú',
    'Uruguay', 'Colón', 'Concordia', 'Federal', 'Feliciano',
    'La Paz', 'Villaguay', 'Nogoyá', 'Tala', 'San Salvador',
    'Federación', 'Islas del Ibicuy'
]

adyacentes = [
    ('Paraná', 'Diamante'), ('Paraná', 'Nogoyá'), ('Paraná', 'Villaguay'), ('Paraná', 'La Paz'),
    ('La Paz', 'Villaguay'), ('La Paz', 'Federal'), ('La Paz', 'Feliciano'),
    ('Diamante', 'Nogoyá'), ('Diamante', 'Victoria'),
    ('Feliciano', 'Federal'), ('Feliciano', 'Federación'),
    ('Victoria', 'Nogoyá'), ('Victoria', 'Gualeguay'),
    ('Nogoyá', 'Victoria'), ('Nogoyá', 'Gualeguay'), ('Nogoyá', 'Tala'), ('Nogoyá', 'Villaguay'),
    ('Villaguay', 'Federal'), ('Villaguay', 'San Salvador'), ('Villaguay', 'Colón'), 
    ('Villaguay', 'Uruguay'), ('Villaguay', 'Tala'),
    ('Federal', 'Federación'), ('Federal', 'Concordia'), ('Federal', 'San Salvador'),
    ('Federación', 'Concordia'),
    ('Concordia', 'San Salvador'), ('Concordia', 'Colón'),
    ('San Salvador', 'Colón'),
    ('Colón', 'Uruguay'),
    ('Uruguay', 'Tala'), ('Uruguay', 'Gualeguaychú'),
    ('Tala', 'Gualeguay'), ('Tala', 'Gualeguaychú'),
    ('Gualeguay', 'Gualeguaychú'), ('Gualeguay', 'Islas del Ibicuy'),
    ('Gualeguaychú', 'Islas del Ibicuy')
]

dominio = {}
colores = ['rojo', 'violeta', 'azul', 'verde']
for departamento in departamentos:
    dominio[departamento] = colores.copy()

def restriccion_diferentes_colores(variables, valores):
    return valores[0] != valores[1]

# Crear restricciones
restricciones = []
for dep1, dep2 in adyacentes:
    restricciones.append(((dep1, dep2), restriccion_diferentes_colores))

problema = CspProblem(departamentos, dominio, restricciones)

solucion = backtrack(problema)

print("Solución encontrada:")
for departamento, color in solucion.items():
    print(f"{departamento}: {color}")

colores_usados = set(solucion.values())
print(f"\nSe utilizaron {len(colores_usados)} colores: {', '.join(colores_usados)}")

# Visualizar el resultado
G = nx.Graph()
G.add_nodes_from(departamentos)
G.add_edges_from(adyacentes)

color_map = {'rojo': 'red', 'verde': 'green', 'azul': 'blue', 'violeta': 'purple'}
node_colors = [color_map[solucion[node]] for node in G.nodes()]

plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color=node_colors)
plt.title("Coloreo del mapa de Entre Ríos")
plt.show()
```

## **Misioneros y caníbales**

Tres misioneros y tres caníbales se encuentran en la orilla de un río. Todos deben cruzar al otro lado utilizando un bote que solo puede llevar a **una o dos personas por vez**. En ningún momento puede haber más caníbales que misioneros, tanto en las orillas como en el bote, ya que los caníbales devorarían a los misioneros.
Hay que encontrar la "configuración" para lograr que los seis individuos crucen el río sanos y salvos.

1. Elegir algún tipo de búsqueda visto en clases para resolver el problema y justificar su elección.

Elegi la busqueda por anchura (BFS):
1. El movimiento de cruzar el rio tiene el mismo costo
2. Si existe una solucion, BFS la puede encontrar
3. BFS explora todos los estados posibles nivel por nivel, no pierdo ninguna solucion
4. Es relativamente facil de implementar para este problema

2. Resuelvan el acertijo utilizando el tipo de búsqueda elegido en el punto anterior. Explicar cada línea de código con comentarios.

**Recuerden**:

Estado incial = todos en la orilla izquierda

Estado objetivo = todos en la orilla derecha

```python
from collections import deque

def resolver_misioneros_canibales():
    # Estado inicial (m_izq, c_izq, posición_bote)
    # m_izq son los misioneros en la orilla izquierda
    # c_izq son los canibales en la orilla izquierda

    # posicion_bote = 0 si esta en la izquierda y 1 si está en la derecha
    estado_inicial = (3, 3, 0)
    
    # El estado objetivo es que todos esten en la orilla derecha
    estado_objetivo = (0, 0, 1)
    
    # Posibles movimientos de los misiones o canibales
    # Un positivo significa que se mueven de izquierda a derecha
    # Un negativo significa que se mueven de derecha a izquierda
    movimientos = [
        (1, 0),  # 1 misionero cruza
        (2, 0),  # 2 misioneros cruzan
        (0, 1),  # 1 caníbal cruza
        (0, 2),  # 2 caníbales cruzan
        (1, 1)   # 1 misionero y 1 caníbal cruzan
    ]
    
    # Una cola para la búsqueda en anchura
    cola = deque([(estado_inicial, [])])  # Pongo el estado inicial y el camino recorrido hasta ahora
    
    # Aca este conjunto almacena los estados visitados y para evitar ciclos
    visitados = set([estado_inicial])
    
    # Mientras haya estados por explorar
    while cola:
        # Tomo el primer estado de la cola
        estado_actual, camino = cola.popleft()
        
        # Si esta en el estado objetivo, se encontro una solucion
        if estado_actual == estado_objetivo:
            return camino
        
        # Saco los valores del estado actual
        m_izq, c_izq, bote = estado_actual
        
        # Calculo los valores de la orilla derecha
        m_der = 3 - m_izq
        c_der = 3 - c_izq
        
        # Pruebo cada posible movimiento
        for mov_m, mov_c in movimientos:
            # Si el bote esta en la izquierda los movimientos son de izquierda a derecha
            if bote == 0:
                nuevo_m_izq = m_izq - mov_m
                nuevo_c_izq = c_izq - mov_c
                nuevo_bote = 1
            # Si el bote esta en la derecha los movimientos son de derecha a izquierda
            else:
                nuevo_m_izq = m_izq + mov_m
                nuevo_c_izq = c_izq + mov_c
                nuevo_bote = 0
            
            # Calculo el nuevo estado de la orilla derecha
            nuevo_m_der = 3 - nuevo_m_izq
            nuevo_c_der = 3 - nuevo_c_izq
            
            # Verifico si el movimiento es valido
            # No puede haber un numero negativo de personas
            if nuevo_m_izq < 0 or nuevo_c_izq < 0 or nuevo_m_der < 0 or nuevo_c_der < 0:
                continue
                
            # No puede haber mas canibales que misioneros en ninguna orilla (si hay misioneros)
            if (nuevo_m_izq > 0 and nuevo_m_izq < nuevo_c_izq) or (nuevo_m_der > 0 and nuevo_m_der < nuevo_c_der):
                continue
                
            # Creamos el nuevo estado
            nuevo_estado = (nuevo_m_izq, nuevo_c_izq, nuevo_bote)
            
            # Si no hemos visitado este estado antes, lo añadimos a la cola
            if nuevo_estado not in visitados:
                visitados.add(nuevo_estado)
                # Agregamos el movimiento al camino
                nuevo_camino = camino + [(mov_m, mov_c, nuevo_bote)]
                cola.append((nuevo_estado, nuevo_camino))
    
    # Si salimos del bucle sin encontrar solución, es que no existe
    return None

# Función para imprimir la solución de manera legible
def imprimir_solucion(solucion):
    if solucion is None:
        print("No hay solución")
        return
        
    print("Solución encontrada:")
    print("Paso 0: Estado inicial")
    print("Orilla izquierda: 3 misioneros, 3 caníbales, bote")
    print("Orilla derecha: 0 misioneros, 0 caníbales")
    print("-----------------------------")
    
    # Estado actual
    m_izq, c_izq = 3, 3
    
    for i, (mov_m, mov_c, bote) in enumerate(solucion):
        # Actualizamos el estado según el movimiento
        if bote == 1:  # Si el bote va a la derecha (desde la izquierda)
            m_izq -= mov_m
            c_izq -= mov_c
            direccion = "derecha"
        else:  # Si el bote va a la izquierda (desde la derecha)
            m_izq += mov_m
            c_izq += mov_c
            direccion = "izquierda"
            
        m_der = 3 - m_izq
        c_der = 3 - c_izq
        
        # Describimos el movimiento
        print(f"Paso {i+1}: Mueven {mov_m} misioneros y {mov_c} caníbales hacia la {direccion}")
        
        # Describimos el estado resultante
        bote_izq = "" if bote == 1 else ", bote"
        bote_der = ", bote" if bote == 1 else ""
        
        print(f"Orilla izquierda: {m_izq} misioneros, {c_izq} caníbales{bote_izq}")
        print(f"Orilla derecha: {m_der} misioneros, {c_der} caníbales{bote_der}")
        print("-----------------------------")

# Ejecutamos la solución
solucion = resolver_misioneros_canibales()
imprimir_solucion(solucion)
```
